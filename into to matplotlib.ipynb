# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt


#1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig ,  ax = plt.subplots(figsize =(10,10))

#3. Plot data
ax.plot(x , y)

#4. Customize plot
ax.set(title = 'Simple Plot',
      xlabel = 'x-axis',
      ylabel = 'y-axis')

#5. Save and show (you save the whole figure)
fig.savefig("sample-plot.png")
```

# Two options for subplots

```python
#Subplot option 1
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize = (10, 5))
#Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2
fig, ax = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize = (10, 5))
#Plot to each different axis
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one should you use?(pyplot or matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more quickly advanced, use the OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
#Pyplot method
over_50.plot(x = 'age' , y = 'chol' ,c = 'target',  kind = 'scatter', figsize = (10,6));
```

```python
#OO method mixed with pyplot method
fig , ax = plt.subplots(figsize = (10,6))
over_50.plot(x ="age", y = "chol" , kind= "scatter", c="target" , ax = ax);
#ax.set_xlim([40,100])
```

```python
over_50.target.values
```

```python
#OO method from scratch
fig , ax = plt.subplots(figsize =(10,6))

#Plot the data
scatter = ax.scatter(x= over_50['age'], y= over_50['chol'], c =over_50['target']);

#customize
ax.set(title ='Heart disease over > 50',
      xlabel = "Age > 50",
      ylabel = 'Cholestrol');

#Add legend
ax.legend(*scatter.legend_elements(), title = 'Target');

#Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
over_50.head()
```

```python
#Subplot of age,chol,thalach
fig , (ax0,ax1) = plt.subplots(nrows =2 ,
                              ncols = 1,
                              figsize = (10,10))
#add data to ax0
scatter = ax0.scatter(x =over_50['age'], y =over_50['chol'], c= over_50['target'])

#customize
ax0.set(title = 'age vs chol',
       xlabel = "age 50",
       ylabel = 'chol');

#legend
ax0.legend(*scatter.legend_elements(), title = 'target');

#horiz line
ax0.axhline(y= over_50['chol'].mean(), linestyle = '--');


#subplot for ages vs thalach
scatter = ax1.scatter(x=over_50['age'], y= over_50['thalach'], c= over_50['target'])
 #customize
ax1.set(title = "age vs thalach",
       xlabel = 'age',
       ylabel = 'thalach')
#legend
ax1.legend(*scatter.legend_elements(), title = 'target')

#vertical line
ax1.axvline(x =over_50['age'].mean(), linestyle = "--");
```

```python
#Subplot of age,chol,thalach
fig , (ax0,ax1) = plt.subplots(nrows =2 ,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)
#add data to ax0
scatter = ax0.scatter(x =over_50['age'], y =over_50['chol'], c= over_50['target'])

#customize
ax0.set(title = 'age vs chol',
       
       ylabel = 'chol');

#legend
ax0.legend(*scatter.legend_elements(), title = 'target');

#horiz line
ax0.axhline(y= over_50['chol'].mean(), linestyle = '--');


#subplot for ages vs thalach
scatter = ax1.scatter(x=over_50['age'], y= over_50['thalach'], c= over_50['target'])
 #customize
ax1.set(title = "age vs thalach",
       xlabel = 'age',
       ylabel = 'thalach')
#legend
ax1.legend(*scatter.legend_elements(), title = 'target')

#vertical line
ax1.axvline(x =over_50['age'].mean(), linestyle = "--");

#add title to the figure
fig.suptitle("heart disease analysis", fontsize = 16 , fontweight = "bold");
```

```python
plt.style.use('seaborn-whitegrid')

#OO method from scratch
fig , ax = plt.subplots(figsize =(10,6))

#Plot the data
scatter = ax.scatter(x= over_50['age'], y= over_50['chol'], c =over_50['target'], cmap = 'winter'); #cmpa-color map and this changes the color scheme

#customize
ax.set(title ='Heart disease over > 50',
      xlabel = "Age > 50",
      ylabel = 'Cholestrol');

#Add legend
ax.legend(*scatter.legend_elements(), title = 'Target');

#Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
#subplot of age,chol,thalach
fig , (ax0,ax1) = plt.subplots(nrows =2 ,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)
#add data to ax0
scatter = ax0.scatter(x =over_50['age'], y =over_50['chol'], c= over_50['target'], cmap ='winter')

#customize
ax0.set(title = 'age vs chol',
       
       ylabel = 'chol');

#changing x axis limit
ax.set_xlim([50,80])

#legend
ax0.legend(*scatter.legend_elements(), title = 'target');

#horiz line
ax0.axhline(y= over_50['chol'].mean(), linestyle = '--');


#subplot for ages vs thalach
scatter = ax1.scatter(x=over_50['age'], y= over_50['thalach'], c= over_50['target'], cmap ="winter")
 #customize
ax1.set(title = "age vs thalach",
       xlabel = 'age',
       ylabel = 'thalach')

#changing x axis limit
ax.set_xlim([50,80])
ax.set_ylim([60,200])
#legend
ax1.legend(*scatter.legend_elements(), title = 'target')

#vertical line
ax1.axvline(x =over_50['age'].mean(), linestyle = "--");

#add title to the figure
fig.suptitle("heart disease analysis", fontsize = 16 , fontweight = "bold");
```
